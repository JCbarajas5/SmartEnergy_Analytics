```python
import pandas as pd
import numpy as np

# Cargar datos
# En esta versión de pandas, parse_dates con diccionario no es compatible
df = pd.read_csv('household_power_consumption.txt', sep=';', na_values=['?'])

# Unir Date y Time en una sola columna datetime
df['datetime'] = pd.to_datetime(
    df['Date'].astype(str) + ' ' + df['Time'].astype(str),
    dayfirst=True,
    errors='coerce'
)
df = df.drop(columns=['Date', 'Time'])

# Limpieza básica
df = df.dropna().set_index('datetime')

# Feature engineering temporal
df['hour'] = df.index.hour
df['day_of_week'] = df.index.dayofweek
df['month'] = df.index.month

# Agregación horaria
df_hourly = df.resample('h').mean()

print(f'Registros originales: {len(df)}')
print(f'Registros horarios: {len(df_hourly)}')
print(df.head())
print(df_hourly.head())
```
